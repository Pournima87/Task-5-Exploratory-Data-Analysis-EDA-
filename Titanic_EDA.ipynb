```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
# Basic Data Exploration
# 3. Get basic information about the dataset
df.info()

# 4. Summary statistics
df.describe()

# 5. Checking for missing values
df.isnull().sum()

# 6. Value counts for important columns
print(df['alive'].value_counts())
print(df['sex'].value_counts())
print(df['pclass'].value_counts())
```

```python
# Data Cleaning (if required)
# 7. Filling missing values in 'Age' with mean
df['age'] = df['age'].fillna(df['age'].mean())


# 8. Dropping the 'Cabin' column due to too many missing values
if 'Cabin' in df.columns:
    df.drop('Cabin', axis=1, inplace=True)
else:
    print("Column 'Cabin' not found in the DataFrame.")

# 9. Filling missing 'Embarked' values with the mode
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
# Verify again
df.isnull().sum()
```

```python
# Drop 'deck' column (because too many missing)
if 'deck' in df.columns:
    df.drop('deck', axis=1, inplace=True)
else:
    print("Column 'deck' not found in the DataFrame, skipping drop.")

# Fill missing 'embark_town' with mode (most frequent value)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)

# Verify again
df.isnull().sum()
```

```python
# Data Visualization
# Histograms for Numeric Variables
# 10. Plot histograms
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
# Boxplots to Detect Outliers
# 11. Boxplot for Age across Passenger Classes
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Age vs Pclass')
plt.show()
```

```python
# Pairplot to See Feature Relationships
# 12. Pairplot
sns.pairplot(df, hue='survived')
plt.show()
```

```python
# Correlation Heatmap
# 13. Correlation matrix
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=np.number)
mask = np.triu(np.ones_like(numeric_df.corr()))
sns.heatmap(numeric_df.corr(), mask=mask, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Barplots for Categorical Variables
# 14. Survival Rate by Gender
ax = sns.barplot(x='sex', y='survived', data=df, hue='sex', palette='pastel', legend=False)
plt.title('Survival Rate by Gender')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

```python
# 15. Survival Rate by Passenger Class
sns.barplot(x='pclass', y='survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```
